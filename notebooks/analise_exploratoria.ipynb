# Média, Moda , médiana e desvio padrão

```python
def obter_resumo_estatistica(dataframe: pd.DataFrame):
    print(f'Média preço: {round(dataframe["precos"].mean(), 2)}')
    print(f'Moda preço: {round(dataframe["precos"].mode(), 2)}')
    print(f'Mediana preço: {round(dataframe["precos"].median(), 2)}')
    print(
        f'Desvio Padrão preço: {round(dataframe["precos"].std(), 2)}')
    print(f'Preço Máximo: {round(dataframe["precos"].max(), 2)}')
    print(f'Preço Minimo: {round(dataframe["precos"].min(), 2)}')
    print(f'Total imóveis: {dataframe.shape}')
```

# Intervalo de confiança

```python
from scipy.stats import norm
```

```python
import numpy as np


def obter_intervalo_de_confianca(df: pd.DataFrame, bairros):
    resultados = []

    for bairro in bairros:
        df_bairro = df.query(f'bairro == "{bairro}"')
        media = df_bairro['precos'].mean()
        desvio_padrao = df_bairro['precos'].std()
        tamanho_amostra = n = df_bairro.shape[0]
        significancia = 0.05
        confianca = 1 - significancia
        raiz_de_n = np.sqrt(n)
        sigma = desvio_padrao / raiz_de_n

        inferior, superior = norm.interval(
            confidence=confianca, loc=media, scale=sigma
        )
        inferior = inferior.round(2)
        superior = superior.round(2)

        resultados.append(
            {'bairro': bairro, 'intervalo_inferior': inferior, 'intervalo_superior': superior})

    return pd.DataFrame(resultados)
```

# Teste de hipótese

- Verificar se os preços dos apartamentos de até 60 m2 são maiores que os apartamentos de 60 metros.

- Apartamento

<ul>
    <li>m1 = média de preços para quartos menores ou iguais que 60 m²</li>
    <li>m2 = média de preços para quartos maiores que 60 m²</li>
    <li>H0: m1 &le; m2 → Hipótese nula</li>
    <li>Ha: m1 &gt; m2 → Hipótese Alternativa</li>
</ul>


```python
from statsmodels.stats.weightstats import DescrStatsW, CompareMeans

def teste_hipotese(base: pd.DataFrame, bairros):
    resultados = []
    for bairro in bairros:
        grupo1 = base[base['metragems']
                      <= 60].sample(n=100, random_state=101).precos
        grupo2 = base[base['metragems'] > 60].sample(
            n=100, random_state=101).precos
        base_ate_60 = DescrStatsW(grupo1)
        base_maior_60 = DescrStatsW(grupo2)
        test_preco = base_ate_60.get_compare(base_maior_60)
        z, p_value = test_preco.ztest_ind(alternative='larger', value=0)
        
        alpha = 0.05
        if p_value < alpha:
            resultado = 'Rejeitamos a hipótese nula.'
            conclusao = 'Há evidências suficientes para afirmar que a média dos preços dos apartamentos de até 60 m² é maior do que a média dos preços dos apartamentos maiores que 60 m²'
        else:
            resultado = 'Não rejeitamos a hipótese nula.'
            conclusao = 'Não há evidências suficientes para afirmar que a média dos preços dos apartamentos de até 60 m² é maior do que a média dos preços dos apartamentos maiores que 60 m².'
        pd.set_option('display.max_rows', None, 'display.max_columns', None)
        resultados.append({'bairro': bairro, 'resultado': resultado, 'conclusao': conclusao})
    return pd.DataFrame(resultados)
```

```python
df_teste_hipotese_ap = teste_hipotese(base_apartamento, bairros_apartamentos_sel)
df_teste_hipotese_ap
```
